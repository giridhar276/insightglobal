## Task 41: Noun phrase chunking

**Problem Statement:** Perform noun phrase chunking using a simple grammar rule.


```python
# Task 41 

# Chunking groups tagged words into higher-level phrases such as noun phrases.

from nltk.tokenize import word_tokenize
from nltk import pos_tag, RegexpParser

sentence = "The talented student won the national competition."
tagged = pos_tag(word_tokenize(sentence))

grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
cp = RegexpParser(grammar)
result = cp.parse(tagged)
print(result)
```

## Task 45: Extract person names

**Problem Statement:** Extract person entities from a sentence using NER.


```python
# Task 45 

# This task focuses only on person-type named entities.

from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

sentence = "Ratan Tata met Narendra Modi in New Delhi."
tagged = pos_tag(word_tokenize(sentence))
tree = ne_chunk(tagged)

persons = []
for subtree in tree:
    if hasattr(subtree, "label") and subtree.label() == "PERSON":
        name = " ".join(token for token, _ in subtree.leaves())
        persons.append(name)

print(persons)
```

## Task 50: Top frequent words in corpus

**Problem Statement:** Find the top 20 frequent words in a selected corpus after removing stopwords.


```python
# Task 50 

# This task combines corpus access, cleaning, stopword removal, and frequency analysis.

import nltk
nltk.download("gutenberg")
from nltk.corpus import gutenberg, stopwords
from nltk import FreqDist

file_id = gutenberg.fileids()[0]
tokens = [w.lower() for w in gutenberg.words(file_id) if w.isalpha()]
stop_words = set(stopwords.words("english"))
clean = [w for w in tokens if w not in stop_words]

fd = FreqDist(clean)
print(fd.most_common(20))
```
