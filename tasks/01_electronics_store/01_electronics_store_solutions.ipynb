## Task 40

Create a new column Gross_Value as Quantity multiplied by Unit_Price.

```python
df['Gross_Value'] = df['Quantity'] * df['Unit_Price']
df[['Quantity','Unit_Price','Gross_Value']].head()
```

## Task 41

Verify whether Gross_Value and Order_Value are consistent after discount calculation.

```python
df['Expected_Order_Value'] = (df['Gross_Value'] * (1 - df['Discount_Percent']/100)).round(2)
(df['Expected_Order_Value'] == df['Order_Value']).value_counts()
```

## Task 42

Create a new column Discount_Amount.

```python
df['Discount_Amount'] = (df['Gross_Value'] - df['Order_Value']).round(2)
df[['Gross_Value','Order_Value','Discount_Amount']].head()
```

## Task 43

Create a new column High_Value_Order where Order_Value > 5000 is Yes otherwise No.

```python
df['High_Value_Order'] = np.where(df['Order_Value'] > 5000, 'Yes', 'No')
df[['Order_Value','High_Value_Order']].head()
```

## Task 44

Use apply() to categorize Unit_Price into Low, Medium, High.

```python
def price_band(x):
    if x < df['Unit_Price'].quantile(0.33):
        return 'Low'
    elif x < df['Unit_Price'].quantile(0.66):
        return 'Medium'
    return 'High'

df['Price_Band'] = df['Unit_Price'].apply(price_band)
df[['Unit_Price','Price_Band']].head()
```
