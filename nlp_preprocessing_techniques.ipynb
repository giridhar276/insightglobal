# NLP Preprocessing Techniques in Python


For each technique, this notebook covers:
- What it means
- Why it is used
- A Python example
- Notes on when to use it

---


## 1. Sample Text

We will use the following sample text across examples.


```python
sample_text = "NLP is amazing! It helps computers understand human language. Running, runs, and ran are related words."
print(sample_text)
```

## 2. Lowercasing

Lowercasing converts all text into lowercase letters.

**Why use it?**
- Reduces variation in text
- Treats `Python` and `python` as the same word

**When to use it?**
Usually used in text classification, search, and basic NLP pipelines.

```python
text = sample_text
lower_text = text.lower()
print(lower_text)
```

## 3. Tokenization

Tokenization means splitting text into smaller units such as words or sentences.

**Why use it?**
- Most NLP tasks work on tokens rather than raw text
- It is the first step before stopword removal, stemming, lemmatization, and vectorization

```python
import nltk
nltk.download('punkt', quiet=True)

from nltk.tokenize import word_tokenize, sent_tokenize

word_tokens = word_tokenize(sample_text)
sentence_tokens = sent_tokenize(sample_text)

print("Word Tokens:", word_tokens)
print("Sentence Tokens:", sentence_tokens)
```

## 4. Removing Punctuation

Punctuation removal deletes symbols like `. , ! ? ; :` from text.

**Why use it?**
- Reduces noise in many NLP tasks
- Helps standardize tokens

**When not to use it?**
In sentiment analysis or chat data, punctuation may carry meaning.

```python
import string

text = sample_text
no_punct = text.translate(str.maketrans('', '', string.punctuation))
print(no_punct)
```

## 5. Removing Numbers

This removes digits from text.

**Why use it?**
- Useful when numbers are not important for the task

**When not to use it?**
Keep numbers when they carry meaning, such as dates, prices, IDs, versions, or measurements.

```python
import re

text = "There are 25 students in NLP class 2026."
no_numbers = re.sub(r'\d+', '', text)
print(no_numbers)
```

## 6. Stopword Removal

Stopwords are very common words such as `is`, `the`, `and`, `of`.

**Why use it?**
- Removes low-information words
- Helps focus on meaningful words

**Caution**
For tasks like question answering or translation, removing stopwords may hurt performance.

```python
import nltk
nltk.download('stopwords', quiet=True)

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
tokens = word_tokenize(sample_text.lower())
filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

print("Original Tokens:", tokens)
print("After Stopword Removal:", filtered_tokens)
```

## 7. Stemming

Stemming reduces words to their root form by chopping off prefixes or suffixes.

Examples:
- `running` → `run`
- `studies` → `studi`

**Why use it?**
- Reduces word variations
- Useful in search and retrieval

**Limitation**
Stemming may produce non-dictionary words.

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()
tokens = word_tokenize("running runs runner easily fairly")
stemmed = [stemmer.stem(word) for word in tokens]

print("Original:", tokens)
print("Stemmed :", stemmed)
```

## 8. Lemmatization

Lemmatization converts words to their dictionary base form.

Examples:
- `running` → `run`
- `better` → `good` (in some systems)

**Why use it?**
- More accurate than stemming
- Produces meaningful root words

**Limitation**
Usually slower than stemming.

```python
import nltk
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemmatizer = WordNetLemmatizer()
tokens = word_tokenize("running runs better cats")
lemmas = [lemmatizer.lemmatize(word) for word in tokens]

print("Original:", tokens)
print("Lemmas  :", lemmas)
```

## 9. Normalization

Normalization means converting text into a standard form.

This can include:
- lowercasing
- removing punctuation
- removing extra spaces
- expanding contractions
- standardizing spellings

```python
import re
import string

text = "  Hello,   NLP World!!  "
text = text.lower()
text = text.translate(str.maketrans('', '', string.punctuation))
text = re.sub(r'\s+', ' ', text).strip()

print(text)
```

## 10. Removing Extra Whitespace

Text often contains multiple spaces, tabs, or newline characters.

**Why use it?**
- Makes text cleaner
- Helps avoid unnecessary token differences

```python
import re

text = "NLP    is\n\nvery\tinteresting."
clean_text = re.sub(r'\s+', ' ', text).strip()
print(clean_text)
```

## 11. Handling Contractions

Contractions are shortened forms like:
- `don't` → `do not`
- `it's` → `it is`

**Why use it?**
- Makes text more explicit
- Helps models understand meaning better

```python
contractions = {
    "don't": "do not",
    "it's": "it is",
    "can't": "cannot",
    "won't": "will not"
}

text = "It's amazing and I can't deny it."
for short, full in contractions.items():
    text = text.replace(short, full).replace(short.capitalize(), full.capitalize())

print(text)
```

## 12. Handling Special Characters

Special characters include symbols such as `@`, `#`, `$`, `%`, `&`, emojis, and HTML tags.

**Why use it?**
- Removes unwanted noise
- Useful in web-scraped or social media text

```python
import re

text = "Hello @user! Welcome to NLP #Python 😊"
clean_text = re.sub(r'[@#]\w+', '', text)   # remove mentions and hashtags
clean_text = re.sub(r'[^\w\s]', '', clean_text)  # remove special chars/emojis
print(clean_text)
```

## 13. Removing HTML Tags

When text comes from websites, it may contain HTML tags.

**Why use it?**
- Helps extract clean readable text from raw HTML

```python
from bs4 import BeautifulSoup

html_text = "<html><body><h1>NLP</h1><p>This is a <b>sample</b> text.</p></body></html>"
clean_text = BeautifulSoup(html_text, "html.parser").get_text(separator=" ")
print(clean_text)
```

## 14. Spelling Correction

Spelling correction fixes misspelled words.

**Why use it?**
- Improves quality of input text
- Useful in search, chatbots, and user-generated content

**Limitation**
Automatic correction may introduce wrong changes.

```python
# Example using TextBlob (install if needed: pip install textblob)
try:
    from textblob import TextBlob
    text = "I havv goood speling in this sentense."
    corrected = str(TextBlob(text).correct())
    print("Original :", text)
    print("Corrected:", corrected)
except Exception as e:
    print("TextBlob not installed or environment issue:", e)
```

## 15. Sentence Segmentation

Sentence segmentation splits a paragraph into sentences.

**Why use it?**
- Useful for summarization, translation, sentiment per sentence, and information extraction

```python
from nltk.tokenize import sent_tokenize

paragraph = "NLP is fun. It is used in chatbots. It also helps in translation."
sentences = sent_tokenize(paragraph)

for i, sentence in enumerate(sentences, 1):
    print(f"Sentence {i}: {sentence}")
```

## 16. N-grams

N-grams are groups of consecutive words.

Examples:
- Unigram: one word
- Bigram: two words
- Trigram: three words

**Why use it?**
- Captures short context and word combinations

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

tokens = word_tokenize("I love learning NLP with Python")
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("Tokens   :", tokens)
print("Bigrams  :", bigrams)
print("Trigrams :", trigrams)
```

## 17. Part-of-Speech Aware Lemmatization

A word's lemma can depend on its part of speech.

Example:
- `running` as a verb → `run`
- `better` as an adjective may map differently

**Why use it?**
- Gives better lemmatization results

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print("Without POS:")
print("running ->", lemmatizer.lemmatize("running"))
print("better  ->", lemmatizer.lemmatize("better"))

print("\nWith POS:")
print("running (verb) ->", lemmatizer.lemmatize("running", pos='v'))
print("better (adj)   ->", lemmatizer.lemmatize("better", pos='a'))
```

## 18. Text Preprocessing Pipeline

In practice, preprocessing is usually applied as a pipeline.

A common order is:
1. lowercasing
2. remove HTML / special characters
3. remove punctuation
4. tokenize
5. remove stopwords
6. stemming or lemmatization

```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text = "NLP is AMAZING!! It helps computers understand human language."

# Step 1: lowercase
text = text.lower()

# Step 2: remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Step 3: tokenize
tokens = word_tokenize(text)

# Step 4: remove stopwords
stop_words = set(stopwords.words('english'))
final_tokens = [word for word in tokens if word not in stop_words]

print("Processed Tokens:", final_tokens)
```

## 20. Final Notes

There is no single preprocessing pipeline that is best for every NLP task.

Choose preprocessing based on your use case:
- **Text classification** → lowercase, tokenize, stopword removal, lemmatization
- **Search / information retrieval** → tokenization, stemming, normalization
- **Sentiment analysis** → be careful with punctuation and negation
- **Chat / social media text** → handle emojis, hashtags, mentions, contractions
