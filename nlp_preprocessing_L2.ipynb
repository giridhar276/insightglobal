# 1. Text Pre-processing L1  
## Tokenization, Lemmatization, Stop Words

This is the **foundation stage** of NLP preprocessing.

Before any model can learn from text, the text has to be cleaned and broken into manageable parts.  
This stage helps standardize raw text.

The three techniques in this level are:
- **Tokenization**
- **Lemmatization**
- **Stop-word removal**


## 1.1 Tokenization

### What is tokenization?
Tokenization means splitting text into smaller units called **tokens**.

A token can be:
- a word
- a sentence
- sometimes a character or subword

### Why is tokenization important?
A machine learning model cannot directly work with a full paragraph as raw text.  
It first needs text to be broken into pieces.

### Types of tokenization
- **Word tokenization** → split into words
- **Sentence tokenization** → split into sentences
- **Subword tokenization** → used by transformer models such as BERT

### Example
Text:
> "NLP is fun. It is powerful."

Sentence tokens:
- NLP is fun.
- It is powerful.

Word tokens:
- NLP
- is
- fun
- .
- It
- is
- powerful
- .


````python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "NLP is fun. It is powerful."

print("Sentence Tokens:")
print(sent_tokenize(text))

print("\nWord Tokens:")
print(word_tokenize(text))
````

### More Example with a Larger Sentence

````python
text = sample_text

word_tokens = word_tokenize(text)
sentence_tokens = sent_tokenize(text)

print("Original Text:")
print(text)

print("\nSentence Tokens:")
print(sentence_tokens)

print("\nWord Tokens:")
print(word_tokens)
````

### Where tokenization is used
- sentiment analysis
- machine translation
- question answering
- search engines
- text summarization
- chatbots


## 1.2 Lemmatization

### What is lemmatization?
Lemmatization converts a word into its **dictionary base form**, called the **lemma**.

### Examples
- running → run
- studies → study
- cars → car

### Why is it useful?
Different forms of a word usually represent the same basic meaning.

For example:
- run
- running
- ran

These should often be treated as related forms.

### Lemmatization vs stemming
**Stemming**
- chops off word endings
- may produce non-real words
- faster but rough

**Lemmatization**
- uses vocabulary and grammar
- gives more meaningful root forms
- usually better for quality


````python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "cars", "studies", "better", "mice"]
for w in words:
    print(f"{w} -> {lemmatizer.lemmatize(w)}")
````

### Why POS matters in lemmatization

Some words need **part-of-speech (POS)** information to get the correct lemma.

For example:
- running as a verb → run
- better as an adjective → good (in some lemmatizers/systems)

Without POS information, lemmatization may be less accurate.


````python
print("Without POS:")
for w in ["running", "better", "driving"]:
    print(f"{w} -> {lemmatizer.lemmatize(w)}")

print("\nWith POS:")
print("running (verb) ->", lemmatizer.lemmatize("running", pos="v"))
print("better (adj)   ->", lemmatizer.lemmatize("better", pos="a"))
print("driving (verb) ->", lemmatizer.lemmatize("driving", pos="v"))
````

### Use cases of lemmatization
- search and retrieval
- document clustering
- topic modeling
- text classification
- text normalization before feature extraction


## 1.3 Stop Words

### What are stop words?
Stop words are very common words that often do not add much useful meaning in many NLP tasks.

Examples:
- the
- is
- and
- are
- in
- on
- of

### Why remove stop words?
They often:
- appear very frequently
- increase noise
- make feature space larger
- add little distinguishing value in some tasks

### But be careful
Stop-word removal is not always a good idea.

Example:
- "not good"

If you remove **not**, the meaning changes completely.

So for:
- sentiment analysis
- question answering
- chatbots
- translation

you should apply stop-word removal carefully.


````python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

text = "The cat is sitting on the mat and looking at the dog."
tokens = word_tokenize(text.lower())
filtered = [w for w in tokens if w.isalpha() and w not in stop_words]

print("Original Tokens:")
print(tokens)

print("\nAfter Stop-word Removal:")
print(filtered)
````

### Example on the sample text

````python
tokens = word_tokenize(sample_text.lower())
filtered = [w for w in tokens if w.isalpha() and w not in stop_words]

print("Tokens:")
print(tokens)

print("\nFiltered Tokens:")
print(filtered)
````

## 1.4 Combining L1 Techniques

A typical basic preprocessing pipeline may look like:
1. lowercase text
2. tokenize
3. remove stop words
4. lemmatize the remaining tokens


````python
text = "The cats are running faster than the dogs in the garden."

tokens = word_tokenize(text.lower())
tokens = [w for w in tokens if w.isalpha()]
tokens_no_stop = [w for w in tokens if w not in stop_words]
lemmas = [lemmatizer.lemmatize(w, pos="v") for w in tokens_no_stop]

print("Lowercased Tokens:", tokens)
print("Without Stop Words:", tokens_no_stop)
print("Lemmatized Tokens:", lemmas)
````

# 2. Text Pre-processing L2  
## Bag of Words, TF-IDF, Unigram, Bigram

After cleaning text, the next step is to convert it into **numbers**.

Machine learning models do not understand raw text directly.  
They need numerical features.

This level focuses on frequency-based representations:
- **Unigram**
- **Bigram**
- **Bag of Words**
- **TF-IDF**


## 2.1 Unigram

A **unigram** is a single token.

Example:
> "I love NLP"

Unigrams:
- I
- love
- NLP

Unigrams are the simplest building blocks in many NLP systems.


````python
text = "I love NLP"
tokens = word_tokenize(text)
print("Unigrams:")
print(tokens)
````

## 2.2 Bigram

A **bigram** is a pair of consecutive tokens.

Example:
> "I love NLP"

Bigrams:
- I love
- love NLP

Bigrams help preserve local word order and short phrases.
This is useful because phrases like:
- not good
- machine learning
- artificial intelligence

carry stronger meaning together.


````python
from nltk.util import ngrams

text = "I love NLP"
tokens = word_tokenize(text)
bigrams = list(ngrams(tokens, 2))

print("Tokens:", tokens)
print("Bigrams:", bigrams)
````

### More example with a phrase-sensitive sentence

````python
text = "This movie is not good"
tokens = word_tokenize(text.lower())

print("Unigrams:")
print(tokens)

print("\nBigrams:")
print(list(ngrams(tokens, 2)))
````

## 2.3 Bag of Words (BoW)

### What is Bag of Words?
Bag of Words converts text into a numeric vector using **word counts**.

### Main idea
- Create a vocabulary from all documents
- Count how many times each word appears in each document

### Why called a bag?
Because word order is ignored.
It only cares about which words are present and how often.

### Example
Documents:
- "I love NLP"
- "I love Python"

Vocabulary:
- I
- love
- NLP
- Python

Vectors:
- [1, 1, 1, 0]
- [1, 1, 0, 1]


````python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

docs = ["I love NLP", "I love Python"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)

feature_names = vectorizer.get_feature_names_out()
df_bow = pd.DataFrame(X.toarray(), columns=feature_names)

print("Vocabulary:")
print(feature_names)

print("\nBag of Words Matrix:")
print(df_bow)
````

### Advantages of Bag of Words
- simple
- interpretable
- fast baseline
- works surprisingly well for basic classification

### Limitations
- ignores word order
- ignores meaning similarity
- sparse representation
- large vocabulary


## 2.4 TF-IDF

### What is TF-IDF?
TF-IDF stands for:
- **TF** = Term Frequency
- **IDF** = Inverse Document Frequency

### Why use TF-IDF?
Bag of Words only counts words.  
But some words appear in almost every document and are not very informative.

TF-IDF gives high importance to words that:
- appear often in a specific document
- but do not appear in every document

### Example intuition
If the word "Python" appears many times in one document but not in all documents, it gets a high weight.
If the word "is" appears everywhere, it gets a low weight.


````python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [
    "I love NLP",
    "I love Python",
    "Python is powerful for NLP"
]

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(docs)

df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())

print("TF-IDF Features:")
print(tfidf.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(df_tfidf.round(3))
````

### Advantages of TF-IDF
- better weighting than raw counts
- highlights important terms
- strong baseline for classification and retrieval

### Limitations
- still does not deeply understand semantics
- usually ignores long-range context
- sparse vectors


## 2.5 Unigram + Bigram with CountVectorizer

We can tell scikit-learn to include both unigrams and bigrams.


````python
docs = [
    "machine learning is fun",
    "I love machine learning",
    "deep learning is powerful"
]

vectorizer = CountVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(docs)

df_ngram = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(df_ngram)
````

## 2.6 Bigram-only Example
Sometimes we only want two-word phrases.


````python
vectorizer = CountVectorizer(ngram_range=(2, 2))
X = vectorizer.fit_transform(docs)

df_bigram_only = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(df_bigram_only)
````

## 2.7 Summary of L2

- **Unigram** = single words
- **Bigram** = two-word sequences
- **Bag of Words** = count-based representation
- **TF-IDF** = importance-weighted representation

These are classic feature engineering methods in NLP.
